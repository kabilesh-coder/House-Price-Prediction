## 3. ***Data Wrangling***

**Handle Missing Values**

- Numerical missing values - Median
- Categorical missing values - None / Mode

```python
# Identify numerical & categorical columns
num_cols = house_df.select_dtypes(include=['int64', 'float64']).columns
cat_cols = house_df.select_dtypes(include=['object']).columns
```

```python
# Handle numerical missing values
for col in num_cols:
    house_df[col].fillna(house_df[col].median(), inplace=True)
```

**Why median?**
>- Median is robust of outliers
>- Mean can be distorted by extreme values

```python
# Handle categorical missing values
for col in cat_cols:
    house_df[col].fillna('None', inplace=True)
```

**Why'None'?**
> Missing often means feature not present
>- No pool
>- No alley
>- no fence

So 'None' becomes a meaningful category, not random data 

**Check the Missing Values After Treatment**

```python
house_df.isnull().sum().sum()
```

**Handle Outliers (Basic)**

```python
# Using GrlivArea
Q1 = house_df['GrLivArea'].quantile(0.25)
Q3 = house_df['GrLivArea'].quantile(0.75)
IQR = Q3 - Q1

upper_limit = Q3 + 1.5 * IQR
house_df['GrLivArea'] = np.where(
    house_df['GrLivArea'] > upper_limit,
    upper_limit,
    house_df['GrLivArea']
)
```

```python
# Drop Irrelevant Column
house_df.drop('Id', axis=1, inplace=True)
```

### What all manipulations have you done and insights you found?

- Data wrangling was performed to prepare the dataset for further analysis and modeling. Missing values in numerical variables were handled using median imputation, while missing values in categorical variables were treated by assigning a ‘None’ category to represent the absence of features. This ensured that no information was lost and maintained data consistency.

- Outliers were handled using the Interquartile Range (IQR) method to reduce the impact of extreme values on model performance. The identifier column ‘Id’ was removed as it does not contribute to prediction. 

- These data wrangling steps improved data quality, reduced noise, and prepared the dataset for visualization, feature engineering, and machine learning modeling.

## ***5. Hypothesis Testing***

### Hypothetical Statement - 1 (OverallQual vs SalePrice)

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

H₀ (Null Hypothesis):
> There is no significant difference in SalePrice across different OverallQual levels.

H₁ (Alternative Hypothesis):
> There is a significant difference in SalePrice across different OverallQual levels.

##### Why did you choose the specific statistical test?

**Test: One-way ANOVA**

**Why ANOVA?**
>- Independent variable: OverallQual (more than 2 groups)
>- Dependent variable: SalePrice (continuous)
>- We want to compare mean prices across multiple quality levels

```python
from scipy.stats import f_oneway

groups = [
    house_df[house_df['OverallQual'] == q]['SalePrice']
    for q in house_df['OverallQual'].unique()
]

f_stat, p_value = f_oneway(*groups)
f_stat, p_value
```

**Conclusion**
>Since the p-value is less than 0.05, the null hypothesis is rejected.
This indicates that house quality has a statistically significant impact on sale price.

### Hypothetical Statement - 2 (GrlivArea vs SalePrice)

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

H₀ (Null Hypothesis):
>There is no significant correlation between GrLivArea and SalePrice.

H₁ (Alternative Hypothesis):
>There is a significant correlation between GrLivArea and SalePrice.

##### Why did you choose the specific statistical test?

**Test: Pearson Correlation Test**

**Why?**
> - Both variables are continuous
>- We want to test linear relationship

```python
from scipy.stats import pearsonr

corr, p_value = pearsonr( house_df['GrLivArea'], house_df['SalePrice'])

corr, p_value
```

**Conclusion**
>The p-value is less than 0.05, indicating a significant positive correlation between living area and sale price.

### Hypothetical Statement - 3 (Neighborhood vs SalePrice)

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

H₀ (Null Hypothesis):
>SalePrice is independent of neighborhood.

H₁ (Alternative Hypothesis):
>SalePrice depends on neighborhood.

##### Why did you choose the specific statistical test?

**Test: One-way ANOVA**

**Why?**
>- Independent variable: Neighborhood (many categories)
>- Dependent variable: SalePrice
>- Comparing means across many locations

```python
groups = [
    house_df[house_df['Neighborhood'] == n]['SalePrice']
    for n in house_df['Neighborhood'].unique()
]

f_stat, p_value = f_oneway(*groups)
f_stat, p_value
```
